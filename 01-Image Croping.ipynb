Harcascade is a machine learning based approach where a cascade function is trained from a lot of positive and negative images. It is then used to detect objects in other images.

Here we will work with face detection. Initially, the algorithm needs a lot of positive images (images of faces) and negative images (images without faces) to train the classifier. Then we need to extract features from it. For this, Haar features shown in the below image are used. They are just like our convolutional kernel. Each feature is a single value obtained by subtracting sum of pixels under the white rectangle from sum of pixels under the black rectangle.

```python
# load haar cascade classifier
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
```

```python
faces = haar.detectMultiScale(gray,1.5,5)
print(faces)
#scalinf factor 1.5 minimum variable 5
```

```python
# Apply to all the images
def extract_images(path,gender,i):
    img = cv2.imread(path)# read image
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # convert to gray scale
    faces = haar.detectMultiScale(gray,1.5,5)# apply harcascade classifier
    for x,y,w,h in faces:# crop image
        roi = img[y:y+h,x:x+w]
        if gender == 'male':# if gender is male as label then put in new folder croppped_image male_crop with following name
            cv2.imwrite('./data/cropped_images/male_crop/{}_{}.png'.format(gender,i),roi)
        else:
            cv2.imwrite('./data/cropped_images/female_crop/{}_{}.png'.format(gender,i),roi)
```

```python
extract_images(male_image[0],'male',1)
```

enumerate

A lot of times when dealing with iterators, we also get a need to keep a count of iterations. Python eases the programmers’ task by providing a built-in function enumerate() for this task.
Enumerate() method adds a counter to an iterable and returns it in a form of enumerate object. This enumerate object can then be used directly in for loops or be converted into a list of tuples using list() method.

```python
for i,path in enumerate(female_image):
    try:

        extract_images(path,'female',i)
        print('INFO: {}/{} processed sucessfully'.format(i,len(female_image)))
        
    except:
        print('INFO: {}/{} cannot be processed'.format(i,len(female_image)))
```

```python
for i,path in enumerate(male_image):
    try:

        extract_images(path,'male',i)
        print('INFO: {}/{} processed sucessfully'.format(i,len(male_image)))
        
    except:
        print('INFO: {}/{} cannot be processed'.format(i,len(male_image)))
```
